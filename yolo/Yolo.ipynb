```python
import cv2
import torch
from ultralytics import YOLO # Install using: pip install ultralytics
# Load a pre-trained YOLOv8 model
model = YOLO('yolov8s.pt')
```

```python
# Task 1: Run YOLO on a sample image
from IPython.display import display, Image
def detect_objects(image_path):
  results= model.predict(image_path, save=True, conf=0.25)
  output_path = f"runs/detect/predict/{image_path.split('/')[-1]}"
  %cd {HOME}
  display(Image(filename=output_path, width=600))
  return results
```

```python
#Parameter
image_path = "https://ultralytics.com/images/zidane.jpg"

output = detect_objects(image_path)
print("Bounding Boxes coordinates for the objects detected are as follows: \n", output[0].boxes.xyxy)
```
